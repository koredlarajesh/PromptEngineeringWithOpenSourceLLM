```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python
model = "meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
def get_response(prompt):
  sequences = pipeline(prompt,
                       do_sample=True,
                       return_full_text=False,
                       top_k=10,
                       num_return_sequences=1,
                       eos_token_id=tokenizer.eos_token_id,
                       max_length=1000)
  return sequences[0]["generated_text"]
```

```python
prompt = '''
<s> [INST] <<sys>>  you are a helpful, respectful and honest assistant, always helpful as answer as possible,
while being a safe, your answers should not be unethical, unsafe , harmful, racisit , sexist , toxic, dangerous, or illegal content.
please ensure that your responses are socially unbiased, gender biased and  positive and nature,
if a question doesn't make sense, or it fatcually coherent, explain why instead of answering something not correct,
if you dont know the answer to a question please dont share false information
<</sys>>
i liked suits in netflix , recommend similar shows to watch
[/INST]
</s>
'''
```

```python
print(get_response(prompt))
```
